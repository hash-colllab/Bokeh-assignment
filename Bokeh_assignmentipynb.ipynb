```python
from bokeh.plotting import figure, output_file,show
from bokeh.sampledata.iris import flowers
```

```python
x=[2,3,4,5,6,8]
y=[5,6,7,8,9,0]
output_file('line.html')
p = figure(title='line plot')
p.line(x,y)
show(p)
```

Glyphs are Bokeh's key building blocks that create plots.

```python
x=[3,4,5,6,7,8,9]
y=[2,3,4,5,6,7,8]
output_file('scatter plot')
p = figure(title='Scatter plot',x_axis_label='x-axis',y_axis_label='y-axis')
p.circle(x,y,size=10,color='blue',alpha=0.5)
show(p)
```

Answer 3:-

```python
x=[1,2,3,4]
y=[3,4,5,6]
output_file('customized_plot.html')
p=figure(title='Customized Plot',x_axis_label='x-axis',y_axis_label='y-axis',plot_width=400,plot_height=300)
p.circle(x,y,size=10,color='blue',alpha=0.5,legend_label='Data Points')
p.title.text_color='pink'
show(p)
```
